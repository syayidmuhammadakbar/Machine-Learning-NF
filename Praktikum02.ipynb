```python
#Visualisasi dengan Histogram

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#ambil data height
data_height = df["Height"]

#membuat hiostogram
n, bins, patches = plt.hist(x=data_height, bins=5, color='pink', edgecolor="black")

#menambahkan label
plt.title("Histogram Nilai")
plt.xlabel("Height")
plt.ylabel("Frekuensi")

#Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins [1:])
plt.xticks(bin_centers, ['{:.1f}-{:.1f}'.format(bins[i], bins[i+1]) for i in range(len(bins)-1)])
#Tampilkan histogram
plt.show()
```

### Subtask:
Use `train_test_split` to split the `mandiri` DataFrame into 80% training data and 20% testing data.

Split the DataFrame into training and testing sets using train_test_split with a test size of 0.2.



**Reasoning**:
The previous attempt failed because `train_test_split` was not imported in the current cell. I need to import it before using it.



```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(mandiri, test_size=0.2, random_state=42)
display(train_df.head())
display(test_df.head())
```

## Split training data into training and validation sets

### Subtask:
Use `train_test_split` again to split the 80% training data into 90% for training and 10% for validation.


**Reasoning**:
Split the existing training data (`train_df`) into new training and validation sets using train_test_split with a test size of 10%.



```python
train_df_final, val_df = train_test_split(train_df, test_size=0.1, random_state=42)
display(train_df_final.head())
display(val_df.head())
```

## Display the shapes of the resulting dataframes

### Subtask:
Show the number of rows and columns in the training, validation, and testing dataframes to verify the split.


**Reasoning**:
Print the shape of the training, validation, and testing dataframes to verify the split proportions.



```python
print("Shape of training data (train_df_final):", train_df_final.shape)
print("Shape of validation data (val_df):", val_df.shape)
print("Shape of testing data (test_df):", test_df.shape)
```
